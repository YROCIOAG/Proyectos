```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
import plotly.express as px
import json
import requests

# 1. Cargar el archivo Excel
ruta_archivo = '/content/Total_2024-2025.xlsx'  # Ruta del archivo
df = pd.read_excel(ruta_archivo)
display("\nCargue Archivo Excel:\n", df)


# 2. Filtro por departamento Cundinamarca
df = df[df['Nombre departamento'] == 'Cundinamarca']
display("\nFiltro por departamento:\n", df)
print(df.columns.tolist()) # Visualizacion de nombres de columnas


# 3. Filtro por Municipios
df = df[df['Nombre municipio '].isin(['Fúquene', 'Guachetá', 'Lenguazaque', 'Simijaca', 'Susa', 'Villa de San Diego de Ubaté'])]
display("\nFiltro por municipios a analizar:\n", df)

# 4. Adicionar coordenadas a los municipios
coordenadas = pd.DataFrame({
    'Nombre municipio ': [
        'Fúquene',
        'Guachetá',
        'Lenguazaque',
        'Simijaca',
        'Susa',
        'Villa de San Diego de Ubaté'
    ],
    'Latitud': [
        5.4000,  # Fúquene
        5.3708,  # Guachetá
        5.2697,  # Lenguazaque
        5.5347,  # Simijaca
        5.4561,  # Susa
        5.3139   # Ubaté
    ],
    'Longitud': [
        -73.7833,  # Fúquene
        -73.6839,  # Guachetá
        -73.6983,  # Lenguazaque
        -73.8119,  # Simijaca
        -73.8506,  # Simijaca
        -73.8147   # Ubaté
    ]
})

# 5. Iguala por municipio para incluir las coordenadas
df = pd.merge(df, coordenadas, on='Nombre municipio ', how='left')


# 6. Ver estructura o descripcion del documento y sus datos de acuerdo a lo adicionado
display(df.head())
display(df.info())
display(df.describe())
display(df.columns.tolist()) # Visualizacion de nombres de columnas

# 7. Limpiar nombres de columnas
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# 8. Conversión fecha para que no genere errores en consultas posteriores
df['mes_y_año'] = pd.to_datetime(df['mes_y_año'])
display("\nConversion Fechas:\n", df)


# 9. Valores duplicados
print("\nRegistros duplicados:\n", df.duplicated().sum())


# 10. Precio promedio por mes
precio_mensual = df.groupby('mes_y_año')['precio_promedio_por_litro'].mean()
precio_mensual.plot(title="Precio promedio mensual por litro", ylabel="Precio ($)", xlabel="Fecha")
plt.grid()
plt.tight_layout()
plt.show()


# 11. Precio promedio por municipio
precio_dep = df.groupby('nombre_municipio')['precio_promedio_por_litro'].mean().sort_values(ascending=False)
display("\nPrecio promedio por municipio:\n", precio_dep)


# 12. Grafica comparativa de municipios y precio promedio
precio_dep.plot(kind='bar', title="Precio promedio por municipio", ylabel="Precio ($)", xlabel="Municipio")
plt.grid()
plt.tight_layout()
plt.show()


# 13. Evolución mensual por municipio (top 3 municipios con mayor precio promedio)
top_municipio = precio_dep.head(3).index
df_top = df[df['nombre_municipio'].isin(top_municipio)]
df_top.groupby(['mes_y_año', 'nombre_municipio'])['precio_promedio_por_litro'].mean().unstack().plot()
plt.title("Evolución mensual del precio por litro (Top 3 municipios)")
plt.ylabel("Precio promedio")
plt.xlabel("Fecha")
plt.grid()
plt.tight_layout()
plt.show()


# 14. Mapa geografico con las coordenadas de los municipios

mapa = folium.Map(location=[5.3, -73.8], zoom_start=9)


# 15. Agregar marcadores al mapa desde el DataFrame
for _, fila in df.iterrows():
    folium.Marker(
        location=[fila['latitud'], fila['longitud']],
        popup=f"{fila['nombre_municipio']}\n${fila['precio_promedio_por_litro']:.2f}",
        tooltip=fila['nombre_municipio']
    ).add_to(mapa)

# 16. Visualización de mapa
display("\nUbicación de municipios en un mapa:\n", mapa)
```
